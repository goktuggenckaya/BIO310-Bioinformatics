### 3 - If a gene’s expression is not varying during the measured time points, it would be hard to identify its role in the process. Therefore, we will filter the genes whose expression do not change over time. Calculate the variance of expression for each gene. (In Python num.var function). Rank the genes based on their variance in descending order (the highest variable gene should be at the top). Which gene varies the most? Which gene varies the least?


```python
sorted_dict = {}
sorted_keys = sorted(gene_ID_to_Var, key=gene_ID_to_Var.get, reverse= True)  # [1, 3, 2]

for w in sorted_keys:
    sorted_dict[w] = gene_ID_to_Var[w]
print(sorted_dict)
count=1
for w in sorted_dict:
  if count == 1:
    most = w
  if count == len(sorted_dict):
    least = w
  count+=1
print("Gene that varies the most is",most, "with variance of", sorted_dict[most],  "while gene that varies the least is", least, "with variance of" , sorted_dict[least])
```

<font color = "indigo">
Above, we rank the genes based on their variance in descending order and found which gene varied the most and which gene varied the least.
